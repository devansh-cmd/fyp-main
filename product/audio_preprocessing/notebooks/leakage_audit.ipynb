```python
import pandas as pd, os, re

def clip_id(p):
    # Adjust regex to your file naming scheme
    return re.sub(r"_(orig|noisy|pitchUp\d+|stretch\d+\.\d+)$", "", os.path.splitext(os.path.basename(p))[0])

train = pd.read_csv("C:/FYP/PROJECT/product/artifacts/splits/train.csv")
val   = pd.read_csv("C:/FYP/PROJECT/product/artifacts/splits/val.csv")

train_ids = set(train["filepath"].apply(clip_id))
val_ids   = set(val["filepath"].apply(clip_id))
overlap   = train_ids & val_ids
print(f"Overlapping clip IDs: {len(overlap)}")
print(list(overlap)[:10])
```

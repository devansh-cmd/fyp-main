```python
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os

AUDIO_DIR = r"C:\FYP\PROJECT\product\audio_preprocessing\data\ESC-50\audio"
SAMPLE_FILES = ["1-100032-A-0.wav", "1-100038-A-14.wav"]
```

```python
def audio_to_mel(y, sr, n_fft=2048, hop_length=512, n_mels=128):
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft,
                                       hop_length=hop_length, n_mels=n_mels) 
    S_db = librosa.power_to_db(S, ref=np.max)  # convert to dB (Log-Mel)
    return S, S_db
```

```python
os.makedirs("product/audio_preprocessing/outputs/spectrograms", exist_ok=True)

for f in SAMPLE_FILES:
    path = os.path.join(AUDIO_DIR, f)
    y, sr = librosa.load(path, sr=None)
    S, S_db = audio_to_mel(y, sr)

    plt.figure(figsize=(12, 4))
    librosa.display.specshow(S_db, sr=sr, hop_length=512, x_axis='time', y_axis='mel', cmap='magma')
    plt.colorbar(format='%+2.0f dB')
    plt.title(f"Log-Mel Spectrogram: {f}")
    plt.tight_layout()

    out_path = f"product/audio_preprocessing/outputs/spectrograms/{f.replace('.wav', '.png')}"
    plt.savefig(out_path, dpi=150, bbox_inches="tight")
    plt.close()
    print(f"Saved spectrogram â†’ {out_path}")
```
